# Using  Yelp API calls

```python
import json
```

```python
with open('/Users/vinosraj/.secret/yelp_api.json') as f: #change the path to match YOUR path!!
    login = json.load(f)
login.keys()
```

```python
from yelpapi import YelpAPI
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# results search
search_results = yelp_api.search_query(location='Rochester, MN',
                                       term='food')
type(search_results)
search_results.keys()
```

```python
search_results['total']
```

```python
import pandas as pd
```

```python
biz = pd.DataFrame(search_results['businesses'])
biz.head(3)
```

```python
#adding an offset of 20
search_results = yelp_api.search_query(location='Rochester, MN',
                                       term='food',
                                      offset = 20)
```

```python
search_results['total']
```

```python
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head(3)
```

```python
biz.head(3)
```

```python
#concatenate the last two results

businesses = pd.concat([biz,biz20],
                      ignore_index= True)

display(businesses.head(3),businesses.tail(3))
```

```python
businesses.to_json(orient='records')
```
